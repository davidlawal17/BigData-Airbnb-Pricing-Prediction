```python
"""Imports"""
from sklearn.model_selection import train_test_split, RandomizedSearchCV, cross_val_score
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
from sklearn.preprocessing import StandardScaler
import numpy as np
import pandas as pd

```

```python
"""Feature Engineering: preparing the dataframe for Random Forest"""

#Creating a new feature: interaction between 'accommodates' and 'bedrooms'
df['accommodates_bedrooms_interaction'] = df['accommodates'] * df['bedrooms']

#Encoding categorical variables (e.g., neighbourhood, property type, and room type)
categorical_vars = ['neighbourhood', 'property_type', 'room_type']
X_cat = pd.get_dummies(df[categorical_vars], drop_first=True)  # One-hot encoding

#Selecting numerical variables (excluding price and log_price as target variables)
num_variables = df.select_dtypes(include=['int64', 'float64']).columns.tolist()
num_variables.remove('price')  # Price is the target variable (log-transformed)
num_variables.remove('log_price')  # Log price is the target variable
num_variables.remove('accommodates_vis')  # Keeping 'accommodates' for visualization"""

#Extracting numerical features
X_num = df[num_variables]

#Combing numerical and categorical features
X_combined = pd.concat([X_num, X_cat], axis=1)

#Scaling the features to standardize them
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_combined)  # Standardization (mean=0, variance=1)

#Defining the target variable: log-transformed price
y = df['log_price']
```

```python
"""Training-test split"""

#Spliting the dataset into training and testing sets (80% train, 20% test)
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
```

```python
"""RandomForestRegressor model"""

rf = RandomForestRegressor(n_estimators=100, random_state=42)

#Fitting the model on the training data
rf.fit(X_train, y_train)

#Making predictions on the test data
y_pred_rf = rf.predict(X_test)

#Evaluating the model performance using R2, RMSE, and MAE
r2_rf = r2_score(y_test, y_pred_rf)
mse_rf = mean_squared_error(y_test, y_pred_rf)  # Mean Squared Error
rmse_rf = np.sqrt(mse_rf)  # Root Mean Squared Error (RMSE)
mae_rf = mean_absolute_error(y_test, y_pred_rf)  # Mean Absolute Error

#Outputing the performance metrics of the Random Forest model
print("\nRandom Forest Model:")
print(f"R: {r2_rf:.4f}")
print(f"RMSE: {rmse_rf:.4f}")
print(f"MAE: {mae_rf:.4f}")
```

```python
"""Hypertuning the Random Forest Model"""

# Define the parameter grid for RandomizedSearchCV
rf_param_dist = {
    'n_estimators': [100, 200, 300],
    'max_depth': [None, 10, 20],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
}

# Instantiate RandomForestRegressor
rf = RandomForestRegressor(random_state=42)

# Perform RandomizedSearchCV to find the best hyperparameters
random_search_rf = RandomizedSearchCV(estimator=rf, param_distributions=rf_param_dist, 
                                      n_iter=10, cv=3, scoring='neg_mean_squared_error', 
                                      random_state=42, n_jobs=-1, verbose=1)

# Fit the RandomizedSearchCV model to the training data
random_search_rf.fit(X_train, y_train)

# Print the best hyperparameters found by RandomizedSearchCV
print("Best Random Forest Hyperparameters:", random_search_rf.best_params_)

# Cross-validation on the best estimator found by RandomizedSearchCV
cv_scores = cross_val_score(random_search_rf.best_estimator_, X_scaled, y, cv=5, scoring='neg_mean_squared_error')

# Calculate RMSE from the cross-validation scores
cv_rmse = np.sqrt(-cv_scores.mean())  # Convert negative MSE to positive and compute RMSE
print(f"Cross-validation RMSE: {cv_rmse:.4f}")
```

```python
 """Testing hypertuned Random Forest Model"""
#Applying the best hyperparameters found during RandomizedSearchCV
best_rf = RandomForestRegressor(n_estimators=200, min_samples_split=5, min_samples_leaf=1, max_depth=None, random_state=42)

#Fitting the model on the entire training data
best_rf.fit(X_train, y_train)

# Predictions on the test data
y_pred_rf = best_rf.predict(X_test)

#Evaluation metrics
r2_rf = r2_score(y_test, y_pred_rf)
rmse_rf = np.sqrt(mean_squared_error(y_test, y_pred_rf))
mae_rf = mean_absolute_error(y_test, y_pred_rf)

#Printing the final model performance
print("\nTuned Random Forest Model:")
print(f"R: {r2_rf:.4f}")
print(f"RMSE: {rmse_rf:.4f}")
print(f"MAE: {mae_rf:.4f}")
```

```python
"""Cross validation on the final improved Random Forest Model"""

#Defining the tuned RandomForestRegressor
best_rf = RandomForestRegressor(n_estimators=200, min_samples_split=5, 
                                min_samples_leaf=1, max_depth=None, random_state=42)

#Performing cross-validation 
cv_scores = cross_val_score(best_rf, X_scaled, y, cv=5, scoring='neg_mean_squared_error', n_jobs=-1)

#Converting the negative MSE to RMSE by taking the square root
cv_rmse = np.sqrt(-cv_scores.mean())  # Convert negative MSE to positive and compute RMSE

#Printing the cross-validated RMSE
print(f"Cross-validated RMSE: {cv_rmse:.4f}")

```
